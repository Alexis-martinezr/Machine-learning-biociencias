```python
columnas = ['INTUBADO', 'NEUMONIA', 'EMBARAZO',
       'HABLA_LENGUA_INDIG', 'DIABETES', 'EPOC', 'ASMA', 'INMUSUPR',
       'HIPERTENSION', 'OTRA_COM', 'CARDIOVASCULAR', 'OBESIDAD',
       'RENAL_CRONICA', 'TABAQUISMO']
```

```python
def graf_barras(df, x):
    # Agrupamos y contamos pacientes
    plt.figure(figsize=(5,5))
    df_g = df.groupby(['defunciones', x])['id'].nunique().reset_index()

    # Graficamos
    sns.barplot(data=df_g, x=x, y='id', hue ='defunciones')
    plt.title(f'Distribución de {x}')
```

```python
def graf_barras_x_dead_rel(df, x, labels_dict=None):
    # Agrupamos y contamos (sin modificar df original)
    df_g = df.groupby([x, 'defunciones'])['id'].nunique().reset_index()

    # Reemplazo de etiquetas SOLO en df_g
    df_g['defunciones'] = df_g['defunciones'].map({2: "Vivo", 1: "Muerto"})
    if labels_dict:
        df_g[x] = df_g[x].map(labels_dict)

    # Normalizamos por cada categoría de x
    df_g["rel"] = df_g.groupby(x)["id"].transform(lambda y: y / y.sum())

    # Graficamos (relativo)
    sns.barplot(data=df_g, x=x, y="rel", hue='defunciones')

    # Formato
    plt.ylabel("Proporción")
    plt.title(f'Distribución relativa de {x}')
    plt.legend(bbox_to_anchor=(1, 0.2), loc='upper left')
    plt.show()
```

```python
for columna in columnas:
    graf_barras_x_dead_rel(data, columna)
```
